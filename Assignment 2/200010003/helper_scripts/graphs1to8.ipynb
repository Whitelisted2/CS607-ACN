````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
def getcontents(filepath):
    f = open(filepath, "r")
    bigarr = []
    for line in f:
        arr = np.array(line.replace("\n", "").split("\t"))
        # print(line)
        bigarr.append(arr)
    bigarr = np.array(bigarr)
    return bigarr
````

### For Graph 1 (INQ)
----

Run ```bash helper_scripts/graph1.sh``` before the following code.

````python
all_data = getcontents("../output/output1.out")
# print(all_data)
B = np.array(all_data[:, 0]).astype(int) # diff lines
N = np.array(all_data[:, 1]).astype(int) # x axis
avg_LU = np.array(all_data[:, -1]).astype(float) # y axis

series1 = {}
for i in B:
    if i not in series1:
        series1[i] = []
x_vals = []
for i in N:
    if i not in x_vals:
        x_vals.append(i)
for i in range(len(avg_LU)):
    series1[B[i]].append(avg_LU[i])

# print(series)

plt.figure()
plt.xlabel("Number of ports ---->")
plt.ylabel("Average Link Utilization ---->")
for i in series1:
    plt.plot(x_vals, series1[i], marker='o', label=str(i))
plt.legend(title="Buffer Size", loc="center right", bbox_to_anchor=(1,0.4))
plt.title("Utilization vs Num(ports) for varying buffer sizes (p=0.5)")
plt.savefig("../images/img1-p5.png")
plt.show()

avg_PD = np.array(all_data[:, -3]).astype(float)
series8 = {}
for i in B:
    if i not in series8:
        series8[i] = []
for i in range(len(avg_PD)):
    series8[B[i]].append(avg_PD[i])
````

so far: 
- series1: dictionary {B1: [vals of LU for diff N], B2: [ ], ...}
